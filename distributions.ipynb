```python
# Import third-party libraries
import numpy as np
import matplotlib.pyplot as plt

# Initialise the random number generator with a fixed seed
rng = np.random.default_rng(42)

# Set the sample size for all distributions
n = 10_000

# Define the dark theme colours
DARK_BG  = '#1e1e1e'
PANEL_BG = '#2d2d2d'
TEXT     = '#e0e0e0'
GRID     = '#3a3a3a'

# Apply the dark theme to all plots
plt.rcParams.update({
    'figure.facecolor':  DARK_BG,
    'axes.facecolor':    PANEL_BG,
    'axes.edgecolor':    TEXT,
    'axes.labelcolor':   TEXT,
    'axes.titlecolor':   TEXT,
    'xtick.color':       TEXT,
    'ytick.color':       TEXT,
    'text.color':        TEXT,
    'grid.color':        GRID,
    'legend.facecolor':  PANEL_BG,
    'legend.edgecolor':  TEXT,
})
```

```python
# Simulate n fair coin flips
coin_flips = rng.integers(0, 2, size=n)

# Count the occurrences of each outcome
counts = np.bincount(coin_flips)

# Create the figure and axes
fig, ax = plt.subplots(figsize=(5, 4))

# Plot the outcome counts as a bar chart
ax.bar(['Tails', 'Heads'], counts, color=['steelblue', 'tomato'], edgecolor=DARK_BG)
ax.set_title('Fair Coin Flip')
ax.set_xlabel('Outcome')
ax.set_ylabel('Count')
ax.set_ylim(0, n * 0.6)

# Annotate each bar with the exact count
for i, c in enumerate(counts):
    ax.text(i, c + 50, str(c), ha='center', color=TEXT)

# Generate the plot
plt.tight_layout()
plt.show()
```

```python
# Define the idealized Bernoulli(0.5) probabilities
ideal_probs = [0.5, 0.5]

# Create the figure and axes
fig, ax = plt.subplots(figsize=(5, 4))

# Plot the theoretical probabilities as a bar chart
ax.bar(['Tails', 'Heads'], ideal_probs, color=['steelblue', 'tomato'], edgecolor=DARK_BG)
ax.set_title('Fair Coin Flip (Idealized)')
ax.set_ylabel('Probability')
ax.set_ylim(0, 0.6)

# Annotate each bar with the exact probability
for i, p in enumerate(ideal_probs):
    ax.text(i, p + 0.01, f'{p:.2f}', ha='center', color=TEXT)

plt.tight_layout()
plt.show()
```

```python
# Simulate n fair dice rolls
dice_rolls = rng.integers(1, 7, size=n)

# Count the occurrences of each face
faces, face_counts = np.unique(dice_rolls, return_counts=True)

# Create the figure and axes
fig, ax = plt.subplots(figsize=(5, 4))

# Plot the face counts as a bar chart
ax.bar(faces, face_counts, color='mediumseagreen', edgecolor=DARK_BG)
ax.set_title('Fair Dice Roll')
ax.set_xlabel('Outcome')
ax.set_ylabel('Count')
ax.set_xticks(faces)
ax.set_ylim(0, n * 0.25)

# Annotate each bar with the exact count
for f, c in zip(faces, face_counts):
    ax.text(f, c + 20, str(c), ha='center', color=TEXT)

# Generate the plot
plt.tight_layout()
plt.show()
```

```python
# Define the idealized discrete uniform probabilities for a six-sided die
ideal_faces = np.arange(1, 7)
ideal_probs = np.full(6, 1 / 6)

# Create the figure and axes
fig, ax = plt.subplots(figsize=(5, 4))

# Plot the theoretical probabilities as a bar chart
ax.bar(ideal_faces, ideal_probs, color='mediumseagreen', edgecolor=DARK_BG)
ax.set_title('Fair Dice Roll (Idealized)')
ax.set_xlabel('Face')
ax.set_ylabel('Probability')
ax.set_xticks(ideal_faces)
ax.set_ylim(0, 0.25)

# Annotate each bar with the exact probability
for f, p in zip(ideal_faces, ideal_probs):
    ax.text(f, p + 0.003, f'{p:.4f}', ha='center', color=TEXT)

plt.tight_layout()
plt.show()
```

```python
# Sample n values from a standard normal distribution
normal_samples = rng.normal(loc=0, scale=1, size=n)

# Create the figure and axes
fig, ax = plt.subplots(figsize=(5, 4))

# Plot the samples as a normalised histogram
ax.hist(normal_samples, bins=50, color='mediumpurple', edgecolor=DARK_BG, density=True)
ax.set_title('Normal Distribution (μ=0, σ=1)')
ax.set_xlabel('Value')
ax.set_ylabel('Density')

# Generate the plot
plt.tight_layout()
plt.show()
```

```python
# Define the x range for the idealized normal PDF
x = np.linspace(-4, 4, 300)

# Compute the standard normal PDF
pdf = np.exp(-x**2 / 2) / np.sqrt(2 * np.pi)

# Create the figure and axes
fig, ax = plt.subplots(figsize=(5, 4))

# Fill under the PDF curve using the same colour as the histogram
ax.fill_between(x, pdf, color='mediumpurple', alpha=0.7)

# Draw the PDF outline
ax.plot(x, pdf, color='mediumpurple', linewidth=2)
ax.set_title('Normal Distribution (Idealized)')
ax.set_xlabel('Value')
ax.set_ylabel('Density')
plt.tight_layout()
plt.show()
```
